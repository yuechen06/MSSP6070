```python
import pandas as pd

file_path = "/content/drive/My Drive/MSSP6070/data/Student Performance Data.csv"
df = pd.read_csv(file_path)
df.head()


```

```python
import re
def transform_student_perf(df):
    pattern = re.compile(r"(math|reading|writing).*score", flags=re.IGNORECASE)
    score_cols = [c for c in df.columns if pattern.search(c)]
    out = df.copy()
    out["average_score"] = out[score_cols].mean(axis=1)
    out["performance_category"] = out["average_score"].apply(
        lambda x: "High" if x >= 85 else ("Medium" if x >= 70 else "Low")
    )
    return out

df_transformed = transform_student_perf(df)
df_transformed.head()

#I use a lambda on average_score to assign High/Medium/Low categories; the function returns a new DataFrame df_transformed
```
